```python
from  transformers import pipeline
```

```python
generator = pipeline('text-generation',model='gpt2', device_map='auto')
```

```python
from torch import nuclear_norm
def autocomplete_sentences(prompt, max_new_tokens=20, num_options=3, temperature=0.7):
    results = generator(
        prompt,
        max_new_tokens=max_new_tokens,
        num_return_sequences=num_options,
        temperature=temperature,
        do_sample=True,
        top_k=20,
        pad_token_id=generator.tokenizer.eos_token_id
        )
    completions = [result['generated_text'].replace(prompt, '').strip()
    for result in results]
    return completions

prompt = input("Enter the sentence:  ")
completions = autocomplete_sentences(prompt)
for i, completion in enumerate(completions, 1):
    print(f"{i}.{prompt} {completion}")
```
